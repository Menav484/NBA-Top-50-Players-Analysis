```python
## games: The number of games the player has played.
## PLAYER: The name of the player.
## GP (Games Played): Total number of games the player has participated in.
## MIN (Minutes): Total minutes played by the player.
## PTS (Points): Total points scored by the player.
## FGM  (Field Goals Made): Number of successful field goals (baskets) made by the player.
## FGA (Field Goals Attempted): Number of field goals the player has attempted.
## FG% (Field Goal Percentage): The percentage of field goals made out of the total field goals attempted.
## 3PM (3-Point Field Goals Made): Number of successful 3-point shots made by the player.
## 3PA (3-Point Field Goals Attempted): Number of 3-point shots the player has attempted.
## 3P% (3-Point Field Goal Percentage): The percentage of 3-point shots made out of the total 3-point shots attempted.
## FTM (Free Throws Made): Number of successful free throws made by the player.
## FTA (Free Throws Attempted): Number of free throws the player has attempted.
## FT% (Free Throw Percentage): The percentage of free throws made out of the total free throws attempted.
## OREB (Offensive Rebounds): Number of rebounds collected by the player on the offensive end of the court.
## DREB (Defensive Rebounds): Number of rebounds collected by the player on the defensive end of the court.
## REB (Rebounds): Total number of rebounds (offensive and defensive) collected by the player.
## AST (Assists): Number of assists made by the player (passing the ball to a teammate in a way that leads to a score).
## STL (Steals): Number of times the player has legally taken the ball away from an opponent.
## BLK (Blocks): Number of times the player has blocked an opponent's shot.
## TOV (Turnovers): Number of times the player has lost possession of the ball to the opposing team.
## EFG% (Effective Field Goal Percentage): Adjusted field goal percentage that takes into account the fact that 3-point field goals are worth more than 2-point field goals.
## TS% (True Shooting Percentage): A measure of shooting efficiency that takes into account field goals, 3-point field goals, and free throws.
## PTS_PG (Points Per Game): Average points scored per game.
## AST_PG (Assists Per Game): Average assists made per game.
## REB_PG (Rebounds Per Game): Average rebounds collected per game.
```

```python
#  IMPORTING IMPORTANT LIBRARIES

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# LOADING THE FILE


file=pd.read_excel(r"C:\Users\rchet\OneDrive\Desktop\SQL PROJECTS\python\data\Book1.xlsx")
```

## DATA CLEANING

```python
file['OREB']=file['OREB'].replace('-','0')
file['OREB']=file['OREB'].astype(int)

file['DREB']=file['DREB'].replace('-','0')
file['DREB']=file['DREB'].astype(int)

```

```python
file['3PA']=file['3PA'].replace('-','0')
file['3PA']=file['3PA'].astype(int)

file['3PM']=file['3PM'].replace('-','0')
file['3PM']=file['3PM'].astype(int)

file['3P%']=file['3P%'].replace('-','0')
file['3P%']=file['3P%'].astype(int)
```

```python
file[['STL','BLK','TOV']]=file[['STL','BLK','TOV']].replace('-','0')
file[['STL','BLK','TOV']]=file[['STL','BLK','TOV']].astype(int)
```

## VISUALISING THE DATA

```python
file['PTS_PG']=file['PTS']/file['GP']
file['AST_PG']=file['AST']/file['GP']
file['REB_PG']=file['REB']/file['GP']
```

```python
top_players = file.nlargest(10, 'REB')[['PLAYER','GP','REB','DREB','OREB']]

top_players
```

```python
#  which player has teh highest no. of rbounds

plt.bar(top_players['PLAYER'],top_players['REB'],color='pink',edgecolor='black',lw=2,width=.6)
plt.grid(True)
plt.xticks(rotation=45,ha='right',fontsize=12);
plt.xlabel('NAME',c='indigo')
plt.ylabel('POINTS',c='indigo')
plt.title('MOST REBOUNDS',c='indigo')
plt.tight_layout()
```

```python
## show the total rebounds and the offensive and defensive rebounds


rebounds_melted = pd.melt(top_players, id_vars=['PLAYER'], value_vars=['OREB', 'DREB', 'REB'], var_name='Rebound Type', value_name='Rebounds')



rebounds_melted['Rebounds']=pd.to_numeric(rebounds_melted['Rebounds'],errors='coerce').fillna(0).astype(int)

plt.figsize=(8,8)

sns.barplot(x='PLAYER', y='Rebounds', hue='Rebound Type', data=rebounds_melted, palette='viridis')


plt.xticks(rotation=45,ha='right');
```

```python
## show top nba scorers and their race

best_points=file.nlargest(10,'PTS')[['PLAYER','PTS','PTS_PG']]

fig, ax = plt.subplots(figsize=(8, 8))
fig.set_facecolor('lightgrey')
ax.set_facecolor('grey')
plt.bar(best_points['PLAYER'],best_points['PTS'],color=['black','black','black','black','black','white','black','black','black','black'])
plt.grid(True,lw=.3)
plt.title('TOP NBA SCORERS',fontweight='bold')
plt.xlabel('PLAYERS',color='black',fontweight='bold')
plt.ylabel('POINTS',color='black',fontweight='bold')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()


```

```python
## show top nba scorers and theor avg points per game


best_points=file.nlargest(10,'PTS')[['PLAYER','PTS','PTS_PG']]


fig, ax1 = plt.subplots(figsize=(12, 8))

fig.set_facecolor('grey')
ax1.set_facecolor('white')

sns.barplot(x='PLAYER',y='PTS',data=best_points,ax=ax1,color='black')
ax1.set_title('TOP NBA SCORERS AND THEIR AVG POINTS PER GAME',fontsize=18,fontweight='bold')
ax1.set_xlabel('PLAYERS',fontsize=13,fontweight='bold')
ax1.set_ylabel('POINTS',fontsize=13,fontweight='bold')
ax1.set_xticklabels(ax1.get_xticklabels(),rotation=45,ha='right')


ax2=ax1.twinx()

sns.lineplot(x='PLAYER',y='PTS_PG',data=best_points,color='grey')
ax2.set_ylabel(ax2.get_ylabel(),fontsize=13,fontweight='bold')


highlight_points = round(best_points.nlargest(10, 'PTS_PG'),2) 
for i, row in highlight_points.iterrows():
    ax2.plot(row['PLAYER'], row['PTS_PG'], 'o',color='grey')  
    ax2.annotate(f'{row["PTS_PG"]}', xy=(row['PLAYER'], row['PTS_PG']), xytext=(5, 5), textcoords='offset points', color='grey', fontweight='bold')

```

```python
##  how many of the free throws  were made by the  most free throws attempted players

best_free_throw=file.nlargest(10,'FTA')[['PLAYER','FTA','FTM']]

def absolute_value(val):
    a = int(val/100.*sum(best_free_throw['FTA']))
    return f'{a}'

fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,12))


ax1.pie(best_free_throw['FTA'],autopct=absolute_value)
ax1.set_title('FREE THROWS ATTEMPTED',fontweight='bold')
ax1.legend(best_free_throw['PLAYER'], loc='upper right', bbox_to_anchor=(.4,1.7))



def absolute_value2(val):
    a = int(val/100.*sum(best_free_throw['FTM']))
    return f'{a}'


ax2.pie(best_free_throw['FTM'],autopct=absolute_value2)
ax2.set_title('FREE THROWS MADE BY THEM',fontweight='bold')
ax2.legend(best_free_throw['PLAYER'],bbox_to_anchor=(.8,1.7))

 

plt.show()
```
